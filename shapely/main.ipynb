```python
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt

# Define a city boundary (not square)
city_boundary = Polygon([(1, 1), (3, 2), (5, 1), (6, 3), (5, 5), (3, 6), (2, 5), (1, 3)])

# Define taxi locations
taxi1 = Point(2, 4)   # inside city
taxi2 = Point(6, 1.5) # outside city

# Plot everything
x, y = city_boundary.exterior.xy
plt.figure(figsize=(6,6))
plt.plot(x, y, color='blue', linewidth=2, label='City Boundary')
plt.scatter(*taxi1.xy, color='green', label='Taxi 1 (inside)', zorder=5)
plt.scatter(*taxi2.xy, color='red', label='Taxi 2 (outside)', zorder=5)
plt.legend()
plt.title("Realistic City Boundary Example")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.axis('equal')
plt.grid(True)
plt.show()

# Check if taxis are inside
print("Taxi 1 inside city:", city_boundary.contains(taxi1))
print("Taxi 2 inside city:", city_boundary.contains(taxi2))
```
