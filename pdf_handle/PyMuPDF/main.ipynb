```python
doc = fitz.open()
page = doc.new_page()
page.insert_text((72, 72), "This is a new PDF page By Sohan ")
doc.save("new.pdf")
doc.close()
```

```python
html_content = """
<html>
  <body>
    <h1 style="color: navy;">Hello from HTML!</h1>
    <p>This is a <b>simple</b> HTML to PDF test using PyMuPDF.</p>
  </body>
</html>
"""

doc = fitz.open()
page = doc.new_page()
page.insert_htmlbox(page.rect, html_content)

doc.save("html_to_pdf.pdf")
doc.close()
```
