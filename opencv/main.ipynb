```python
import cv2

image = cv2.imread("Facedetection.jpg", cv2.IMREAD_ANYCOLOR)
print(image.shape)

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
print(gray.shape)

_,threshold = cv2.threshold(gray,121,65535, cv2.THRESH_BINARY)

print(threshold.shape)

cv2.imshow("gray Image", gray)
cv2.imshow("threshold Image", threshold)
cv2.imshow("Original Image", image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('road_image.jpg')

cv2.line(image, (50,100), (50,100), (255,0,0), 3)
 
cv2.imshow("Image", image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('road_image.jpg')
cv2.circle(image,(500,100), 30, (255,0,0), -1) 
cv2.imshow("Image", image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

##Rectangle

```python
image = cv2.imread('road_image.jpg')
cv2.rectangle(image,(10,50),(50,300),(0,0,0), -1)

cv2.imshow("Image", image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cam = cv2.VideoCapture(0)
while True:
    ret, frame = cam.read()
    
    if not ret:
        break
    
    cv2.imshow("My came",frame)
    
    if cv2.waitKey(1) & 0xFF == ord("w"):
        break
    
cam.release()
cv2.destroyAllWindows()
```

```python
cam = cv2.VideoCapture(0)

frame_width = int(cam.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cam.get(cv2.CAP_PROP_FRAME_HEIGHT))

codec = cv2.VideoWriter_fourcc(*"XVID")
recorder = cv2.VideoWriter("self_video.avi", codec, 20, (frame_width, frame_height))

while True:
    
    success, frame = cam.read()
    
    if not success :
        break
    
    recorder.write(frame)
    cv2.imshow("My cam", frame)
    
    if cv2.waitKey(1) & 0xFF == ord("w"):
        break

cam.release()
cv2.destroyAllWindows()
```
